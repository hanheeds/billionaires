```python
df = pd.read_csv('../input/billionaires-statistics-dataset/Billionaires Statistics Dataset.csv')
df.head()
```

```python
df['birthDate'] = pd.to_datetime(df['birthDate'])

# Then, extract the year (or other desired integer-valued feature)
df['birthYear'] = df['birthDate'].dt.year

# Now, you can calculate the mean of the years
mean_birth_year = df['birthYear'].mean()

print(f"The average birth year is: {mean_birth_year}")
```
